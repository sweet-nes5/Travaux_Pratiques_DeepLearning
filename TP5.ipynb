## Exercice 1

```python
import torch 
import torch.nn.functional as F
import pub_preliminaries as pb
import matplotlib.pyplot as plt
import itertools

def generate_data_L2C2(nb_samples):
    data_inputs = torch.rand(nb_samples, 2) *2 -1 ** 2
    weight1 = 5
    weight2 = -1
    intercept = -0.8
    bruit = torch.normal(mean=0.0, std=0.1, size=(nb_samples,))
    data_labels = (weight1* (data_inputs[:, 0])**2 + weight2 * data_inputs[:,1] + intercept + bruit) < 0
    return data_inputs, data_labels

data_inputs, data_labels = generate_data_L2C2(1000)

data_labels = torch.cat( (data_labels, torch.zeros(50)))
data_labels = torch.cat( (data_labels, torch.ones(50)))
data_inputs = torch.cat((data_inputs, torch.rand(100, 2) *2 -1))
pb.show_data_L2C2(data_inputs, data_labels)
```

## Exercice 2

```python
s = torch.randperm(1100) # random permutation
labels_shuffled = data_labels[s]
inputs_shuffled = data_inputs[s]
# adding the data to the training set
training_inputs = inputs_shuffled[:880]
training_labels = labels_shuffled[:880]
#adding the data to the validation set
valid_inputs = inputs_shuffled[880:]
valid_labels = labels_shuffled[880:]
pb.show_data_L2C2(training_inputs, training_labels)
pb.show_data_L2C2(valid_inputs, valid_labels)
```

## Exercice 3

```python
model_a = torch.nn.Sequential( torch.nn.Linear(2,10),
                              torch.nn.ReLU(),
                              torch.nn.Linear(10,10),
                              torch.nn.ReLU(),
                              torch.nn.Linear(10,10),
                              torch.nn.ReLU(),
                              torch.nn.Linear(10,1),
                              torch.nn.Sigmoid())

model_b = torch.nn.Sequential( torch.nn.Linear(2,5),
                              torch.nn.ReLU(),
                              torch.nn.Linear(5,10000),
                              torch.nn.ReLU(),
                              torch.nn.Linear(10000,5),
                              torch.nn.ReLU(),
                              torch.nn.Linear(5,1),
                              torch.nn.Sigmoid())

model_c = torch.nn.Sequential( torch.nn.Linear(2,100),
                              torch.nn.ReLU(),
                              torch.nn.Linear(100,10),
                              torch.nn.ReLU(),
                              torch.nn.Linear(10,100),
                              torch.nn.ReLU(),
                              torch.nn.Linear(100,1),
                              torch.nn.Sigmoid())

optimizer_a = torch.optim.SGD(model_a.parameters(), lr=0.1)
optimizer_b = torch.optim.SGD(model_b.parameters(), lr=0.1)
optimizer_c = torch.optim.SGD(model_c.parameters(), lr=0.1)
```

```python
def accuracy(predictions, ys):
    ys = ys.unsqueeze(dim=1) #  ATTENTION: to work with linear
    pred_classes = predictions > 0.5 # fix a threshold 
    success_rate = (pred_classes.to(int) == ys).to(float).mean() # probability of having everything right
    return (success_rate * 100).item()
    

def loss(predictions, ys):
    ys = ys.unsqueeze(dim=1) # ATTENTION: to work with linear
    return (-(ys * torch.log(predictions) + (1-ys) * torch.log(1 - predictions))).mean()
    
def grad_des(epoch, model, optim):
    history = []
    for ep in range(epoch):
        pred = model(training_inputs)
        err = loss(pred, training_labels)
        err.backward()
        optim.step()
        model.zero_grad()

        if ep % 100 == 0:
            acc = accuracy(pred, training_labels)
            loss_val = loss(model(valid_inputs), valid_labels)
            acc_val = accuracy(model(valid_inputs), valid_labels)
            history.append((ep, err.detach().numpy(), acc, loss_val.detach().numpy(), acc_val))
    return history
    
# Plot monitoring
def ith(l, i):
    return [x[i] for x in l]

def plot_history(h):
    fig, axs = plt.subplots(1, 2)
    axs[0].plot(ith(h,0), ith(h,1))
    axs[0].plot(ith(h,0), ith(h,3), color='red')
    axs[0].set_xlabel('epochs')
    axs[0].set_ylabel('loss')
    axs[1].plot(ith(h,0), ith(h,2))
    axs[1].plot(ith(h,0), ith(h,4), color='red')
    axs[1].set_xlabel('epochs')
    axs[1].set_ylabel('accuracy')
    fig.suptitle('Monitor learning')
    fig.tight_layout()
    plt.show()

    
history = grad_des(2000, model_a, optimizer_a)
plot_history(history)
history = grad_des(2000, model_b, optimizer_b)
plot_history(history)
history = grad_des(2000, model_c, optimizer_c)
plot_history(history)
```

## Exercice 4

```python
def generate_data_NL2C3(nb_samples, noise=0.0): 
    """
        returns a pair of tensors (inputs,labels) 
        inputs of shape (nb_samples*3, 3)
        labels of shape (nb_samples*3)
        (Code from http://cs231n.stanford.edu)
    """
    inputs = torch.empty(nb_samples * 3, 2)
    labels = torch.empty(nb_samples * 3, dtype=int)
    for j in range(3):  # number of categories
        ix = range(nb_samples * j, nb_samples * (j + 1))
        r = torch.linspace(0.0, 1, nb_samples)
        t = torch.linspace(j * 4, (j + 1) * 4, nb_samples) + torch.randn(nb_samples) * noise  
        inputs[ix] = torch.column_stack((r * torch.sin(t), r * torch.cos(t)))
        labels[ix] = j
    
    return (inputs, labels)
inputs, labels = generate_data_NL2C3 (1000, 0.6)
pb.show_data_L2C3(inputs, labels)
```

```python
s = torch.randperm(inputs.size(0)) # random permutation
labels_shuff= labels[s]
inputs_shuff = inputs[s]
# adding the data to the training set
train_inputs = inputs_shuff[:2400]
train_labels = labels_shuff[:2400]
#adding the data to the validation set
validation_inputs = inputs_shuff[2400:]
validation_labels = labels_shuff[2400:]
pb.show_data_L2C3(train_inputs, train_labels)
pb.show_data_L2C3(validation_inputs, validation_labels)
```
